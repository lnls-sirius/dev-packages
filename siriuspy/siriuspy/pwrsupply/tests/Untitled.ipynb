```python
%matplotlib notebook
import matplotlib.pyplot as plt
import matplotlib.gridspec as mpl_gs
import matplotlib.cm as cm
import matplotlib.dates as mdates
from matplotlib import rcParams
rcParams.update({'font.size': 16,'lines.linewidth':2})

import numpy as np
import datetime
from siriuspy.clientarch import ClientArchiver as CA


def plot_wfm(time, data):
    data = np.array(data)

    f  = plt.figure(figsize=(9, 5))
    gs = mpl_gs.GridSpec(1, 1)
    gs.update(left=0.10, right=0.9, top=0.9, bottom=0.15, hspace=0.4)
    ax1 = plt.subplot(gs[0, 0])

    indcs = np.linspace(0, 1, len(data))
    cmap = cm.winter(indcs)
    for i, datum in enumerate(data):
        dt = '{:.3f}'.format(time[i]-time[0])
        ax1.plot(datum, '.', color=cmap[i], markersize=1)
    # ax1.legend(bbox_to_anchor=(1.2, 0.5), loc='center right', ncol=1, fontsize='xx-small')
    tit = 'Time interval: {:.3f} s, NrCurves: {:d}'.format(time[-1]-time[0], len(data))
    ax1.grid()
    ax1.set_title(tit)
    ax1.set_xlabel('Wfm index')
    ax1.set_ylabel('Current Diff [A]')
    f.show()
```

```python
ca = CA()
```

```python
ca.login('ximenes.resende', 'boo500mev')
```

```python
psname1 = 'BO-Fam:PS-B-2'
pvname11 = psname1 + ':Wfm-SP'
pvname12 = psname1 + ':Wfm-RB'
pvname13 = psname1 + ':WfmRef-Mon'
pvname14 = psname1 + ':Wfm-Mon'

ini_time = datetime.datetime(2019, 10, 11, 18, 23, 0)
end_time = ini_time + datetime.timedelta(seconds=10)
tstamp1 = ini_time.isoformat() + '-03:00'
tstamp2 = end_time.isoformat() + '-03:00'

info11 = ca.getData(pvname11, tstamp1, tstamp2)
info12 = ca.getData(pvname12, tstamp1, tstamp2)
info13 = ca.getData(pvname13, tstamp1, tstamp2)
info14 = ca.getData(pvname14, tstamp1, tstamp2)
time11, data11 = info11['timestamp'], info11['value']
time12, data12 = info12['timestamp'], info12['value']
time13, data13 = info13['timestamp'], info13['value']
time14, data14 = info14['timestamp'], info14['value']
```

```python
time = time14
data = np.array(data14)
data = data

f  = plt.figure(figsize=(9, 5))
gs = mpl_gs.GridSpec(1, 1)
gs.update(left=0.10, right=0.9, top=0.9, bottom=0.15, hspace=0.4)
ax1 = plt.subplot(gs[0, 0])

# data2 = np.array(data11)
# ax1.plot(datum - 0*data[0], '.', color=cmap[i], markersize=1)

indcs = np.linspace(0, 1, len(data))
cmap = cm.winter(indcs)
for i, datum in enumerate(data):
    dt = '{:.3f}'.format(time[i]-time[0])
    ax1.plot(datum - 0*data[0], '.', color=cmap[i], markersize=1)
# ax1.legend(bbox_to_anchor=(1.2, 0.5), loc='center right', ncol=1, fontsize='xx-small')
tit = 'Time interval: {:.3f} s, NrCurves: {:d}'.format(time[-1]-time[0], len(data))
ax1.grid()
ax1.set_title(tit)
ax1.set_xlabel('Wfm index')
ax1.set_ylabel('Current Diff [A]')
f.show()
```
